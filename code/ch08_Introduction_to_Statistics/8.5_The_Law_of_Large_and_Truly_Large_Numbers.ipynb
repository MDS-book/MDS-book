## 8.5 The Law of Large and Truly Large Numbers

### 8.5.1 The Law of Large Numbers

The law of large numbers (LLN) is one of the important concepts in probability
and statistics. It states that if a random experiment is repeated “sufficiently” many
times and all results are averaged, we will obtain a result that is approximately the
expected value.

The following Python code demonstrates this and roughly reproduces the figure from the book:

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()
```

Obtain 1000 random values and use the cumulative summation to obtain
the means for all lengths of random number sequences:

```python
n_samples = 1000
values = rng.integers(1, 6, endpoint=True, size=n_samples)
sample = np.arange(1, n_samples + 1)
averages = values.cumsum() / sample
```

Plot the deviations of the mean values from the expected value of 3.5:

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 3.5))
ax[0].plot(sample, values, 'o', ms=3, mfc='C0', mec='none', alpha=0.6)
ax[0].set(xlabel="number of rolling a die, $N$", ylabel="value of die rolling")

ax[1].semilogx(sample, averages - 3.5, '-')
ax[1].axhline(0, ls=':')
ax[1].set(ylim=(-2.5, 2.5), xlabel="number of rolling a die, $N$", ylabel="deviation from the mean")
```

We observe that for an increasing number of random draws the sample increases in size, and its mean approaches more and more the expected value, i.e., the population mean.
